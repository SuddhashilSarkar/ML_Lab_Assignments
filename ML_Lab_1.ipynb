```python
#Create a dataframe with 3 columns- Name, Age, Rating
# a. find sum of column
# b. Find avg of age and rating
# c. Find Standard Deviation
# d. Describe the dataframe

import pandas as pd
import numpy as np

data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 28],
        'Rating': [4.5, 3.8, 4.2]}

df = pd.DataFrame(data)

# a. find sum of column
print("Sum of each column:")
print(df.sum())

# b. Find avg of age and rating
print("\nAverage age:", df['Age'].mean())
print("Average rating:", df['Rating'].mean())

# c. Find Standard Deviation
print("\nStandard deviation of age:", df['Age'].std())
print("Standard deviation of rating:", df['Rating'].std())

# d. Describe the dataframe
print("\nDescription of the dataframe:")
print(df.describe())
```

```python
# Consider the following data frame and illustrate application aggregate functions both row-wise and column-wise.
#  A B C
# 0 1.0 2 3.0
# 1 4.0 5 6.0
# 2 7.0 8 9.0
# 3 NaN NaN NaN
# a)Display 1st 10 rows of the data frame. Ans df[:10]
# b)Apply aggregate functions row-wise
# c)Apply aggregate function column-wise.

import pandas as pd
import numpy as np

# Create the DataFrame
data = {'A': [1.0, 4.0, 7.0, np.nan],
        'B': [2, 5, 8, np.nan],
        'C': [3.0, 6.0, 9.0, np.nan]}
df = pd.DataFrame(data)

# a) Display first 10 rows (or all rows if there are less than 10)
print("First 10 rows:")
print(df[:10])

# b) Apply aggregate functions row-wise
df['Row_Sum'] = df.apply(np.sum, axis=1)
df['Row_Mean'] = df.apply(np.mean, axis=1)
print("\nDataFrame with row-wise sum and mean:")
print(df)

# c) Apply aggregate functions column-wise
print("\nColumn-wise sum:")
print(df.sum())

print("\nColumn-wise mean:")
print(df.mean())
```

```python
# prompt: /content/phone_data.csv  use this file
# write
# code to find:
# a. How many rows are there in the dataset?
# b. What was the longest phone call /data entry?
# c. How many seconds of phone calls are recorded in total?
# d. Number of nonnull unique network entries.
# e. How many entries are there for each month?
# f. Get the first entry for each month g. Get the sum of the durations per month

import pandas as pd

df = pd.read_csv('/content/phone_data.csv')

# a. How many rows are there in the dataset?
num_rows = df.shape[0]
print("Number of rows:", num_rows)

# b. What was the longest phone call /data entry?
longest_call = df['duration'].max()
print("\nLongest phone call (seconds):", longest_call)

# c. How many seconds of phone calls are recorded in total?
total_duration = df['duration'].sum()
print("\nTotal duration of phone calls (seconds):", total_duration)

# d. Number of nonnull unique network entries.
unique_networks = df['network'].nunique()
print("\nNumber of unique networks:", unique_networks)

# e. How many entries are there for each month?
entries_per_month = df['month'].value_counts()
print("\nEntries per month:")
print(entries_per_month)

# f. Get the first entry for each month
first_entry_per_month = df.groupby('month').first()
print("\nFirst entry for each month:")
print(first_entry_per_month)

# g. Get the sum of the durations per month
sum_duration_per_month = df.groupby('month')['duration'].sum()
print("\nSum of durations per month:")
print(sum_duration_per_month)
```

```python
# h. Get the number of dates/entries in each month
entries_per_month = df['month'].value_counts()
print("\nNumber of entries per month:")
print(entries_per_month)

# i. Sum of durations, for calls only, to each network
sum_call_duration_per_network = df[df['item'] == 'call'].groupby('network')['duration'].sum()
print("\nSum of call durations per network:")
print(sum_call_duration_per_network)

# j. How many calls, sms, and data entries are in each month?
item_counts_per_month = df.groupby(['month', 'item'])['date'].count().unstack(fill_value=0)
print("\nNumber of calls, sms, and data entries per month:")
print(item_counts_per_month)

# k. How many calls, texts, and data are sent per month, split by network type?
item_counts_per_month_network = df.groupby(['month', 'network', 'item'])['date'].count().unstack(fill_value=0)
print("\nNumber of calls, texts, and data per month, split by network type:")
print(item_counts_per_month_network)
```

```python
# /content/asiangamestop10.csv
#  Related asiangamestop10.csv dataset
# a. Total Number of Records
# b. Top n Records .
# c. Rename ColName 'NOC' to 'Country'
# d. Top 15 Year Wise Medals Count Using BarGraph
# e. Top 15 Counrty Wise Medals Count Using BarGraph
# f. Max, Min and Mean for Total Medals

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/asiangamestop10.csv')

# a. Total Number of Records
num_records = df.shape[0]
print("Total number of records:", num_records)

# b. Top n Records (let's say n=5)
top_5_records = df.head(5)
print("\nTop 5 records:")
print(top_5_records)

# c. Rename ColName 'NOC' to 'Country'
df = df.rename(columns={'NOC': 'Country'})
print("\nDataFrame with renamed column:")
print(df.head())

# d. Top 15 Year Wise Medals Count Using BarGraph
top_15_years = df.groupby('Year')['Total'].sum().nlargest(15)
plt.figure(figsize=(12, 6))
top_15_years.plot(kind='bar')
plt.title('Top 15 Years by Total Medals')
plt.xlabel('Year')
plt.ylabel('Total Medals')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# e. Top 15 Country Wise Medals Count Using BarGraph
top_15_countries = df.groupby('Country')['Total'].sum().nlargest(15)
plt.figure(figsize=(12, 6))
top_15_countries.plot(kind='bar')
plt.title('Top 15 Countries by Total Medals')
plt.xlabel('Country')
plt.ylabel('Total Medals')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# f. Max, Min and Mean for Total Medals
max_medals = df['Total'].max()
min_medals = df['Total'].min()
mean_medals = df['Total'].mean()
print("\nMax Total Medals:", max_medals)
print("Min Total Medals:", min_medals)
print("Mean Total Medals:", mean_medals)
```
