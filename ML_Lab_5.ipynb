```python
import pandas as pd

# Define the data
data = {
    'Date': [
        '10-01-16', '10-02-16', '10-03-16', '10-04-16', '10-05-16',
        '10-06-16', '10-07-16', '10-08-16', '10-09-16', '10-10-16'
    ],
    'Open': [
        770.25, 776.030029, 774.25, 776.030029, 779.3541998,
        776.306698, 779.659973, 776.309998, 775.554009, 776.432029
    ],
    'High': [
        776.075902, 778.730022, 776.065702, 771.710022, 782.073407,
        782.46398, 779.659973, 782.070007, 785.42138, 778.730022
    ],
    'Low': [
        769.5, 780.890015, 766.5, 772.890015, 775.650024,
        775.539978, 770.75, 775.650024, 775.54478, 780.890015
    ],
    'Close': [
        782.659998, 776.429993, 772.549798, 776.429993, 776.469971,
        776.859985, 775.080017, 776.469971, 776.859985, 776.429993
    ]
}

# Create the DataFrame
df = pd.DataFrame(data)

# Convert 'Date' to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y')

# Extract month and year
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

# Example: Filter data for a specific month and year
month_to_filter = 10  # October
year_to_filter = 2016
filtered_df = df[(df['Month'] == month_to_filter) & (df['Year'] == year_to_filter)]

# Display the DataFrame with the new columns and filtered data
print("DataFrame with Month and Year columns:")
print(df)
print("\nFiltered DataFrame for October 2016:")
print(filtered_df)
```

```python
# prompt: Write a Python program to visualize the data from the filtered_df
# a) Line Chart , b) Bar Graph c) Histogram
# show legends and different colours.

import matplotlib.pyplot as plt

# a) Line Chart
plt.figure(figsize=(10, 6))
plt.plot(filtered_df['Date'], filtered_df['Open'], label='Open', color='blue')
plt.plot(filtered_df['Date'], filtered_df['Close'], label='Close', color='red')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Open and Close Prices for October 2016')
plt.legend()
plt.grid(True)
plt.show()

# b) Bar Graph
plt.figure(figsize=(10, 6))
plt.bar(filtered_df['Date'], filtered_df['High'] - filtered_df['Low'], label='Daily Range', color='green')
plt.xlabel('Date')
plt.ylabel('Price Range')
plt.title('Daily Price Range for October 2016')
plt.legend()
plt.grid(True)
plt.show()

# c) Histogram
plt.figure(figsize=(10, 6))
plt.hist(filtered_df['Close'], bins=10, label='Close Prices', color='purple', alpha=0.7)
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.title('Distribution of Closing Prices for October 2016')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# prompt: Read all columns of the filtered_df and show it using a multiline plot

import matplotlib.pyplot as plt
# Plot all columns of the filtered DataFrame
plt.figure(figsize=(10, 6))
for column in filtered_df.columns:
  if column != 'Date':
    plt.plot(filtered_df['Date'], filtered_df[column], label=column)

plt.xlabel('Date')
plt.ylabel('Value')
plt.title('All Columns of Filtered DataFrame')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# prompt: Read the Open column values from dataframe of each month and show it using a scatter
# plot

import pandas as pd
import matplotlib.pyplot as plt

# Extract the 'Open' column values
open_values = df['Open']

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Date'], open_values, label='Open', color='blue', marker='o')
plt.xlabel('Date')
plt.ylabel('Open Price')
plt.title('Scatter Plot of Open Prices')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# prompt: Read Open and Close columns values from filtered_df of each month and plot it dide by side in a barchart with different legends and colours

import pandas as pd
import matplotlib.pyplot as plt

# ... (Preceding code remains the same)

# Group the data by month and extract Open and Close values
monthly_data = filtered_df.groupby('Month')[['Open', 'Close']].mean()

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(monthly_data.index - 0.2, monthly_data['Open'], width=0.4, label='Open', color='blue')
plt.bar(monthly_data.index + 0.2, monthly_data['Close'], width=0.4, label='Close', color='red')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.title('Average Open and Close Prices for October 2016')
plt.xticks(monthly_data.index, ['October'])  # Set x-axis labels to month names
plt.legend()
plt.grid(True)
plt.show()
```

```python
# prompt: Read the High and Low columns values from filtered_df of all months and show it using a
# bar chart

import matplotlib.pyplot as plt
# Group the data by month and extract High and Low values
monthly_data = df.groupby('Month')[['High', 'Low']].mean()

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(monthly_data.index - 0.2, monthly_data['High'], width=0.4, label='High', color='blue')
plt.bar(monthly_data.index + 0.2, monthly_data['Low'], width=0.4, label='Low', color='red')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.title('Average High and Low Prices by Month')
plt.xticks(monthly_data.index, ['October'])  # Set x-axis labels to month names
plt.legend()
plt.grid(True)
plt.show()
```

```python
# prompt: Read the Close column values of each month and display using proper labels :
# a) Histogram to see the most common profit ranges
# b) BarChart to see the most common profit ranges
# c) Pie Chart
# Use filtered_df

import matplotlib.pyplot as plt
# a) Histogram
plt.figure(figsize=(10, 6))
plt.hist(filtered_df['Close'], bins=10, label='Close Prices', color='skyblue', edgecolor='black')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.title('Distribution of Closing Prices for October 2016')
plt.legend()
plt.grid(True)
plt.show()

# b) Bar Chart
close_counts = filtered_df['Close'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(close_counts.index, close_counts.values, color='lightcoral', edgecolor='black')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.title('Frequency of Closing Prices for October 2016')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# c) Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(close_counts.values, labels=close_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Proportion of Closing Prices for October 2016')
plt.show()
```

```python
# prompt: For the Open Columns Data from the filtered_df and show it using a Pie chart.

import matplotlib.pyplot as plt
# Extract the 'Open' column values from the filtered DataFrame
open_values = filtered_df['Open']

# Count the frequency of each open price
open_counts = open_values.value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(open_counts.values, labels=open_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Proportion of Open Prices for October 2016')
plt.show()

```
