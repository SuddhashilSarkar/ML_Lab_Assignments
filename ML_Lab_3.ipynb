```python
# prompt: Create a dataframe as shown below:
# df = pd.DataFrame(data={'a':np.random.randint(0, 100, 30),
# 'b':np.random.randint(0, 100, 30),
# 'c':np.random.randint(0, 100, 30)})
# a) Create a figure of size 15x8 with two subplots, top and bottom.
# Draw two lines in the top axes, one green and one orange.
# Add a legend for the top plot, Green and Orange. Put this legend in the top-middle of graph.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame(data={'a':np.random.randint(0, 100, 30),
                        'b':np.random.randint(0, 100, 30),
                        'c':np.random.randint(0, 100, 30)})

fig = plt.figure(figsize=(15, 8))

ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

ax1.plot(df['a'], color='green', label='Green')
ax1.plot(df['b'], color='orange', label='Orange')
ax1.legend(loc='upper center')
```

```python
# prompt: Challenge: In the bottom axes, create a graph of only the data points, marked by circles, but with no line
# connecting the points.
# b) Draw Bar graph for the above data.

import matplotlib.pyplot as plt
ax2.plot(df['c'], marker='o', linestyle='None', color='purple', label='Purple')
ax2.legend(loc='upper center')

plt.show()

df.plot(kind='bar', figsize=(10, 6))
plt.show()
```

```python
# According to the dataframe:
# a)Create a figure with three subplots, aligned vertically.
# b)Create a bar graph for a and b on the first two,
# c) Create a stacked bar graph by putting b on top of a
# d) Design a piechart for a and b

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.DataFrame(data={'a':np.random.randint(0, 100, 30),
                        'b':np.random.randint(0, 100, 30),
                        'c':np.random.randint(0, 100, 30)})

fig = plt.figure(figsize=(15, 15))

ax1 = fig.add_subplot(311)
ax2 = fig.add_subplot(312)
ax3 = fig.add_subplot(313)


df.plot(kind='bar', y='a', ax=ax1)
df.plot(kind='bar', y='b', ax=ax2)
df[['a', 'b']].plot(kind='bar', stacked=True, ax=ax3)

plt.show()

df[['a', 'b']].sum().plot(kind='pie', figsize=(10, 6))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

Language = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# a) Bar chart with different colors
plt.bar(Language, popularity, color=['red', 'black', 'green', 'blue', 'yellow', 'cyan'])
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("Popularity of Programming Languages")
plt.show()

# b) Grouped bar chart
X = np.arange(3) # Changed this line
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, popularity[0:2], color = 'b', width = 0.25)
ax.bar(X + 0.25, popularity[2:4], color = 'g', width = 0.25)
ax.bar(X + 0.50, popularity[4:6], color = 'r', width = 0.25)
ax.set_xticks(X)
ax.set_xticklabels(['Group 1', 'Group 2', 'Group 3']) # Changed this line
ax.legend(labels=['Group 1', 'Group 2', 'Group 3'])
plt.show()

# c) Stacked bar chart
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(Language, popularity[0:2], color = 'b')
ax.bar(Language, popularity[2:4], bottom = popularity[0:2], color = 'g')
ax.bar(Language, popularity[4:6], bottom = popularity[0:4], color = 'r')
plt.show()
```
